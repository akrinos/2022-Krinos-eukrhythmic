```python
## READ IN ABUNDANCE DATA FROM HARRIET'S 2015 PAPER ##

harriet_tax_info=pd.read_csv("../metadata/NB_HA_MMETSP_140513_Cluster.tab",sep="\t",header=None,
                            names=["Source_ID","HA_lab"])
harriet_tax_reads=pd.read_csv("../metadata/NB_HA_SummedSpecies.tab",sep="\t",
                             header=None,names=["HA_lab","S1","S2","S3","S4","S5",
                                                "A","B","C","D","E"])

harriet_tax_info["Source_ID"] = [curr.strip() for curr in harriet_tax_info["Source_ID"]]
harriet_tax_info["HA_lab"] = [curr.strip() for curr in harriet_tax_info["HA_lab"]]

## MMETSP taxonomy table
taxonomy_mmetsp=pd.read_csv("/vortexfs1/omics/alexander/akrinos/2021-remodeling-eukrhythmic/WAP_utilities/eukulele-workflow/marmmetsp_phaeo_final/tax-table.txt",sep="\t")
merged_taxonomy=pd.merge(harriet_tax_info,taxonomy_mmetsp,left_on="Source_ID",right_on="Source_ID",how="left")


harriet_nb = pd.merge(harriet_tax_reads,merged_taxonomy,how="left",left_on=["HA_lab"],
                      right_on=["HA_lab"])
```

## Look at diatoms at the genus level in Harriet's dataset

```python
distinct_cols=["#bd6c90",
"#77bb3c",
"#c35abc",
"#5dc47a",
"#d04177",
"#4d8c38",
"#7977c2",
"#a6ad5d",
"#5a9dd6",
"#ca4d3f",
"#4ec2b9",
"#ce814a",
"#3c8963",
"#786d29","black"]
diatom_sum_HA = harriet_nb[[str(curr).strip() == "Ochrophyta" for \
                                 curr in harriet_nb.Division]].\
                 groupby("Genus")[["S1","S2","S3","S4","S5"]].sum().\
                 reset_index().\
                 melt(id_vars="Genus")
diatom_sum_HA["GenusCondense"] = ["Other" if numreads < 500000 else genus for genus,numreads in \
                              zip(diatom_sum_HA.Genus,diatom_sum_HA.value)]
(plotnine.ggplot(diatom_sum_HA) +plotnine.theme_bw(base_size=15)+
     plotnine.theme(axis_text_x=plotnine.element_text(rotation=90, hjust=1)) + 
     plotnine.geom_bar(plotnine.aes(x="variable",y="value",fill="GenusCondense"),stat="identity") + 
    plotnine.scale_fill_manual(values=distinct_cols) + plotnine.ylab("reads assigned") + 
    plotnine.xlab("Sample") + plotnine.ggtitle("Harriet's diatom read assignment breakdown"))
```

```python
harriet_nb["ClassificationPlot"] = ["Bacteria" if (str(domain).strip() == "Bacteria")\
                                        else "Archaea" if (str(domain).strip() == "Archaea")\
                                    else "Bacillariophyta-Stramenopiles" if ((str(class_curr).strip() == "Bacillariophyta")|\
                                                        (str(class_curr).strip()=="Stramenopiles") |
                                                        (str(phy).strip()=="Stramenopiles_X")) &\
                                    (str(genus).strip() not in ["Skeletonema","Thalassiosira"]) \
                                    else "Ciliophora" if ((str(phy).strip()=="Ciliophora") | \
                                                          (str(phy).strip()=="Cercozoa")) &\
                                                          #| \
                                                          #(str(phy).strip()=="Apicomplexa")) &\
                                    (str(genus).strip() not in ["Skeletonema","Thalassiosira"]) \
                                    #&
                                    #(phy != "Ochrophyta")\
                                    else phy if (str(genus).strip() not in ["Skeletonema","Thalassiosira"]) \
                                    else genus for phy,genus,domain,class_curr in \
                                    zip(harriet_nb.Division,harriet_nb.Genus,
                                        harriet_nb.Domain,harriet_nb.Class)]
green_list = ["Skeletonema","Thalassiosira","Ciliophora","Haptophyta","Chlorophyta",
              "Ochrophyta","Dinoflagellata","Bacillariophyta-Stramenopiles","Archaea","Bacteria"]
harriet_nb["ClassificationPlotRed"] = ["Other" if (str(classin).strip() not in green_list) \
                                    else classin for classin in harriet_nb.ClassificationPlot]
```

```python
grouped_stacked = harriet_nb.groupby("ClassificationPlotRed")[["S1","S2","S3","S4","S5"]].sum().reset_index().\
    melt(id_vars="ClassificationPlotRed")
grouped_stacked["numerictime"] = [float(curr.split("S")[-1]) for curr in grouped_stacked.variable]
grouped_stacked.value = grouped_stacked.value.astype(float)
grouped_stacked = pd.merge(grouped_stacked,
         grouped_stacked.groupby("variable")["value"].sum().reset_index()\
         .rename(columns={"value":"total"}),
         left_on="variable",right_on="variable",how="left")
grouped_stacked["Proportion"] = grouped_stacked["value"] / grouped_stacked["total"]
```

```python
EUKulele_frame = pd.DataFrame()
EUKulele_frame_full = pd.DataFrame()
output_dir="/vortexfs1/omics/alexander/akrinos/2021-testing-eukrhythmic/eukrhythmic-eLife/snakemake-workflows/eukulele-snake"
subdir="EUKulele_marmmetsp_NB_CAG_contigs"

output_dir="/vortexfs1/omics/alexander/akrinos/2021-testing-eukrhythmic/eukrhythmic_paper_narragansett"
subdir="EUKulele_merged_CAG_6Mar22"
salmon_dir="/vortexfs1/omics/alexander/akrinos/2021-testing-eukrhythmic/"+\
    "eukrhythmic_paper_narragansett/intermediate-files/04-compare/09-CAG-mapping/salmon"

for curr_folder in os.listdir(os.path.join(output_dir)): #,"EUKulele_marmmetsp_NB_CAG")):
    if curr_folder != subdir:
        continue
                                           #'EUKulele_assemblers')):
    curr_assembler = curr_folder.split("EUKulele_")[-1]
    for curr_file in os.listdir(os.path.join(output_dir,subdir,#'EUKulele_assemblers',
                                             #curr_folder,
                                             "taxonomy_estimation")):
        curr_sample = curr_file.split("_" + curr_assembler)[0]
        curr_sample = curr_file.split("_merged")[0] #("_CAG")[0]
        EUKulele_file = pd.read_csv(os.path.join(output_dir,
                                                 subdir,
                                                 #'EUKulele_assemblers',curr_folder,
                                                 "taxonomy_estimation",curr_file),sep="\t")
        EUKulele_file["split_tscpt_name"] = EUKulele_file["transcript_name"]
        #[curr.split(".p")[0] for curr in EUKulele_file["transcript_name"]]
        #[curr.split(curr_sample + "_")[-1] for curr in \
        #                                     EUKulele_file["transcript_name"]]
        salmon_file = pd.read_csv(os.path.join(salmon_dir,
                                               curr_sample + "_quant",
                                               #curr_sample+"_"+curr_assembler+"_quant",
                                               "quant.sf"),sep="\t")
        EUKulele_file = pd.merge(EUKulele_file,salmon_file,left_on="split_tscpt_name",right_on="Name",
                                 how = "right")
        EUKulele_file.loc[EUKulele_file.full_classification != EUKulele_file.full_classification,
                          "full_classification"] = "Unclassified;Unclassified;Unclassified;" + \
                    "Unclassified;Unclassified;Unclassified;Unclassified;Unclassified"
        class_labels = ["Domain","Supergroup","Phylum","Class","Order",
                    "Family","Genus","Species"]
        EUKulele_file[class_labels]= EUKulele_file["full_classification"].\
                                            str.split(";", n = 8, expand = True)
        for class_label in class_labels:
            EUKulele_file.loc[[curr is None for curr in EUKulele_file[class_label]],class_label] = "Unknown"
        EUKulele_file_full = EUKulele_file.copy(deep=True)
        EUKulele_file_full["Sample"] = curr_sample
        EUKulele_file_full["Assembler"] = curr_assembler
        EUKulele_file = EUKulele_file.groupby(["Order","Domain",
                                               "Phylum","Class","Family","Genus",
                                               "Species"]).agg({'TPM': ['count','sum'],
                                                                'NumReads':['count','sum']}).reset_index()
        EUKulele_file.columns = ["".join(col) for col in EUKulele_file.columns.values]
        EUKulele_file["Sample"] = curr_sample
        EUKulele_file["Assembler"] = curr_assembler
        EUKulele_frame = pd.concat([EUKulele_frame,EUKulele_file])
        EUKulele_frame_full = pd.concat([EUKulele_frame_full,EUKulele_file_full])
     
```
